```python
import requests
url="https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/hyderabad/2023-12-01/2023-12-07?unitGroup=metric&elements=datetime%2CdatetimeEpoch%2Ctempmax%2Ctempmin%2Ctemp%2Cfeelslikemax%2Cfeelslikemin%2Cfeelslike%2Chumidity%2Cwindspeed%2Cwindspeedmax%2Cwindspeedmean%2Cwindspeedmin%2Cwinddir%2Csunrise%2Csunset&include=days&key=TWJVW7H78FPMH3S5WGLH7P7K6&contentType=json"
api_link=requests.get(url).json()
print(api_link)
```

```python
import matplotlib.pyplot as plt
```

# TEMP

```python
plt.figure(figsize=(8,5))
plt.xlabel=("date")
plt.ylabel=("temp")
plt.title=("temp as on date")
plt.plot(datetime,temp)
plt.show()
```

# sunrise and sunset

```python
plt.figure(figsize=(8,4))
plt.subplot(2,1,1)
plt.plot(datetime,sunrise)
plt.figure(figsize=(8,4))
plt.subplot(2,1,2)
plt.plot(datetime,sunset)
plt.show()
```

# humidity

```python
plt.figure(figsize=(8,6))
plt.plot(datetime,humidity)
plt.show()
```

# windspeed

```python
plt.figure(figsize=(8,6))
plt.plot(datetime,windspeed)
plt.show()
```

# max temp , min temp and over all temp

```python
plt.figure(figsize=(8,6))
plt.plot(datetime,tempmax,ls='--')
plt.plot(datetime,temp)
plt.plot(datetime,tempmin,ls=':')
plt.show()
```

# whats the temp and how its feels

```python
plt.figure(figsize=(8,6))
plt.plot(datetime,temp)
plt.plot(datetime,feelslike,ls='--')
plt.show()
```
